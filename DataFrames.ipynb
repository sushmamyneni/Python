```python
#CATEGORICAL AND DUMMY VARIABLES
# Categorical variables are those which take a limited usually fixed number of possible values.

raw_data = {'Day': ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday',
                    'Sunday', 'Monday', 'Tuesday', 'Wednesday'],
            'Sex': ['male', 'female', 'male', 'female', 'female', 'female', 'female',
                    'male', 'male', 'female'],
            'Total_Amount': ['100.0', '30', '40', '70', '90', '20', '50', '30',
                             '60', '70']}

#create a dataframe from raw_data
import pandas as pd
raw_data_DF = pd.DataFrame(raw_data,columns=['Day','Sex','Total_Amount'])

# Create a dataframe with columns of type categorical variables
raw_data_DF1 = pd.DataFrame(raw_data,columns=['Day','Sex','Total_Amount'],dtype="category")
raw_data_DF1


# CAtegorical variables examples: Ex gender, bloodtype, daysofweek
# check the type of columns in the datatype raw_data_DF.dtypes

#raw_data_DF['Day'][2:5]

# Why do we need categorical variables? 

#############################################
# Dummy Variables: certain machine learning algorithms can not make sense of categorical variables such as day or sex. We 
# need to transform them to numbered codes to represent such unique values. To do so we can use dummy variables.
# dummy variables transform the day of the week to a value

day_dummies = pd.get_dummies(raw_data_DF['Day'])
day_dummies
```
